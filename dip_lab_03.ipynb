```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt
```

```python
src = cv2.imread("/content/drive/MyDrive/dip_lab/clrpncl.jfif", cv2.IMREAD_UNCHANGED)
DF = cv2.edgePreservingFilter(src, flags=1, sigma_s=60, sigma_r=0.4)
print("Domain Filter - Cartoonify")
cv2_imshow(np.hstack((src, DF)))
```

```python
dst = cv2.GaussianBlur(src,(5,5),cv2.BORDER_DEFAULT)
print("Gaussian Smoothing")
cv2_imshow(np.hstack((src, dst)))
```

```python
kernel = np.ones((10,10),np.float32)/25   
dst2 = cv2.filter2D(src,-1,kernel)
print("Mean Filter")
cv2_imshow(np.hstack((src, dst2)))
```

```python
#Median Filter
dst3 = cv2.medianBlur(src,5)
print("Median Filter")
cv2_imshow(np.hstack((src, dst3)))
```

```python
#Bilateral filter 
print("Bilateral Filter")
dst4 = cv2.bilateralFilter(src, 60, 60, 60) 
cv2_imshow(np.hstack((src, dst4)))
```

```python
#low pass filter
Lp = cv2.filter2D(src,-1, kernel)
Lp = src - Lp
print("Low Pass")
cv2_imshow(np.hstack((src, Lp)))
```

```python
Hp = src - dst
filtered = Hp + 127*np.ones(src.shape, np.uint8)
print("High Pass")
cv2_imshow(np.hstack((src, Hp)))
```

```python
from google.colab.patches import cv2_imshow
kernel = np.ones((10,10),np.float32)/25
meanFilter = cv2.filter2D(src,-1,kernel)
cv2_imshow(np.hstack((src, meanFilter)))
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
domainFilter = cv2.edgePreservingFilter(src, flags=1, sigma_s=60, sigma_r=0.6)
cv2_imshow(np.hstack((domainFilter, Hp)))
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# import required libraries
import numpy as gfg
import matplotlib.image as img
 
# read an image
imageMat = img.imread("/content/drive/MyDrive/dip_lab/colour.jfif")
print("Image shape:", imageMat.shape)
 
# if image is colored (RGB)
if(imageMat.shape[2] == 3):
   
  # reshape it from 3D matrice to 2D matrice
  imageMat_reshape = imageMat.reshape(imageMat.shape[0],
                                      -1)
  print("Reshaping to 2D array:",
        imageMat_reshape.shape)
 
# if image is grayscale
else:
  # remain as it is
  imageMat_reshape = imageMat
     
# saving matrice to .csv file
gfg.savetxt("/content/drive/MyDrive/dip_lab/test.csv",imageMat_reshape)
 
# retrieving matrice from the .csv file
loaded_2D_mat = gfg.loadtxt("/content/drive/MyDrive/dip_lab/test.csv")
 
# reshaping it to 3D matrice
loaded_mat = loaded_2D_mat.reshape(loaded_2D_mat.shape[0],
                                   loaded_2D_mat.shape[1] // imageMat.shape[2],
                                   imageMat.shape[2])
 
print("Image shape of loaded Image:",
      loaded_mat.shape)
 
# check if both matrice have same shape or not
if((imageMat == loaded_mat).all()):
   
  print("\n\nYes",
        "The loaded matrice from CSV file is same as original image matrice")
```
