```python

import cv2
import numpy as np
import matplotlib.pyplot as plt
# Read the image
image_original = cv2.imread('/content/drive/MyDrive/dip_lab/minimal.jpg')
image = cv2.resize(image_original, (400, 400))
# Define the structuring element for erosion and dilation
kernel = np.ones((5, 5), np.uint8)
# Apply erosion
eroded_image = cv2.erode(image, kernel, iterations=1)
# Apply dilation
dilated_image = cv2.dilate(image, kernel, iterations=1)
# Subtract eroded image from the original to get the eroded edge image
eroded_edge_image = cv2.absdiff(image, eroded_image)
# Subtract dilated image from the original to get the dilated edge image
dilated_edge_image = cv2.absdiff(image, dilated_image)
# Display the images using matplotlib
plt.figure(figsize=(12, 8))
plt.subplot(221)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.subplot(222)
plt.imshow(eroded_image, cmap='gray')
plt.title('Eroded Image')
plt.subplot(223)
plt.imshow(eroded_edge_image, cmap='gray')
plt.title('Eroded Edge Image')
plt.subplot(224)
plt.imshow(dilated_edge_image, cmap='gray')
plt.title('Dilated Edge Image')
plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
# Read the image
image_original = cv2.imread('/content/drive/MyDrive/dip_lab/minimal.jpg')
image = cv2.resize(image_original, (400, 400))
# Apply filters for edge detection
edges = cv2.Canny(image, 100, 200)
# Apply filters for texture extraction
blur = cv2.blur(image, (5, 5))
laplacian = cv2.Laplacian(image, cv2.CV_64F)
sobel_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)
# Display the images
plt.figure(figsize=(12, 8))
plt.subplot(231)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.subplot(232)
plt.imshow(edges, cmap='gray')
plt.title('Edges')
plt.subplot(233)
plt.imshow(blur, cmap='gray')
plt.title('Blur')
plt.subplot(234)
plt.imshow(laplacian, cmap='gray')
plt.title('Laplacian')
plt.subplot(235)
plt.imshow(sobel_x, cmap='gray')
plt.title('Sobel X')
plt.subplot(236)
plt.imshow(sobel_y, cmap='gray')
plt.title('Sobel Y')
plt.tight_layout()
plt.show()
```
